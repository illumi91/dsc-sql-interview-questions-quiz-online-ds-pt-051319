## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:

We can have 7 different Joins:
1. Inner -->  Join just elements in common between tables
2. Inner Left -->  Join elements in common between tables all all the elements of the initial table
3. Inner Right -->  Join elements in common between tables all all the elements of the joining table
4. Outer -->  Join all elements not common between tables
5. Outer Left -->  Join just elements not in common between from the initial table
6. Outer Right -->  Join just elements not in common between from the joining table
7. Full Outer -->  Join all elements
